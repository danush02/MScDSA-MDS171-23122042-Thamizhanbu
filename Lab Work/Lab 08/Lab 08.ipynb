# Lab 08 - Matrix Multiplication

```python
mat_dict={"matrix1":[],"matrix2":[],"result":[]}
def collectMatrix1():
    rownum1=int(input("Matrix1-Enter the row size:"))
    columnum1=int(input("Matrix1-Enter the column size:"))
    matrix1=[]
    for i in range(0,rownum1):
            temp_mat=[]
            for j in range(0,columnum1):
                temp_mat.append(int(input(f"Enter the {i+1} row and {j+1} column value of matrix 1")))
            matrix1.append(temp_mat)
    action=input("Shall we store this matrix y/n:")
    if action=="y":
        mat_dict["matrix1"]=matrix1
        print("Matrix 1 stored successfully!\nMatrix 1:")
        for i in mat_dict["matrix1"]:
            print(i)
        return rownum1,columnum1
    elif action=="n":
        collectMatrix1()

def collectMatrix2():  
    rownum2=int(input("Matrix2-Enter the row size:"))
    columnum2=int(input("Matrix2-Enter the column size:"))
    matrix2=[]
    for i in range(0,rownum2):
        temp_mat=[]
        for j in range(0,columnum2):
            temp_mat.append(int(input(f"Enter the {i+1} row and {j+1} column value of matrix 2")))
        matrix2.append(temp_mat)
    action=input("Shall we store this matrix y/n:")
    if action=="y":
        mat_dict["matrix2"]=matrix2
        print("Matrix 2 stored successfully!\nMatrix 2:")
        for i in mat_dict["matrix2"]:
            print(i)
        return rownum2,columnum2
    elif action=="n":
        collectMatrix2()
    
def matrixMultiplication(rownum2,columnum2):
    for i in mat_dict["matrix1"]:
        temp_mat=[]
        for l in range(0,columnum2):
            j=0
            sum=0
            for k in range(0,rownum2):
                sum+=i[j]*mat_dict["matrix2"][k][l]
                j+=1
            temp_mat.append(sum)
        mat_dict["result"].append(temp_mat)
    print("Matrix Calculated Succesfully!\n")
    for i in mat_dict["result"]:
            print(i)

def storeToFile():
    file_mat=open("resultMatrix.csv","w+")
    for i in mat_dict["result"]:
        temp=len(i)
        check=1
        for j in i:
            if temp==1:
                file_mat.write(str(j)+"\n")
            elif temp<=check:
                file_mat.write(str(j)+",")
                check+=1
            else:
                file_mat.write(str(j)+"\n")
    file_mat.close()
    print("File stored successfully!")

            
```

```python
while True:
    print("Please select the action\n1.Collect Matrix Value\n2.Calculate Matrix Multiplication\n3.Export Result to File\n4.Display Result from File\n5.Exit")
    action=int(input("Enter your choice:"))
    if action==1:
        rownum1,columnum1=collectMatrix1()
        rownum2,columnum2=collectMatrix2()
        if columnum1!=rownum2:
            print("Matrix can't be multiplied. Please check the matrix.")
    elif action==3:
        if not mat_dict["matrix1"]:
            print("Please set matrix 1")
        elif not mat_dict["matrix2"]:
            print("Please set matrix 2")
        else:
            matrixMultiplication(rownum2,columnum2)
    elif action==4:
        if not mat_dict["result"]:
            print("Please multiply the matrices before saving.")
        else:
            storeToFile()

        
```
